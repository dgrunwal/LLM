```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
```

```python
(X_train, y_train), (X_test,y_test) = cifar10.load_data()
```

```python
# prompt: Generate a function called display_cifar that can show a specified number of images and their class name

import matplotlib.pyplot as plt
import numpy as np
def display_cifar(num_images):
    # Define the class names
    class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

    # Create a figure and axes for the subplots
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))

    # Loop through the specified number of images
    for i in range(num_images):
        # Select a random index from the training data
        random_index = np.random.randint(0, len(X_train))

        # Get the image and label at the random index
        image = X_train[random_index]
        label = y_train[random_index][0]  # Access the label value (0-9)

        # Display the image on the corresponding subplot
        axes[i].imshow(image)
        axes[i].set_title(class_names[label])
        axes[i].axis('off')

    # Adjust spacing and display the plot
    plt.tight_layout()
    plt.show()
```

```python
display_cifar(5)
```
